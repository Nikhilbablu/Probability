```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
N = 365
customer_visits = np.clip(np.random.normal(loc=5, scale=2, size=N).astype(int), 0, 10)

unique_values, counts = np.unique(customer_visits, return_counts=True)
pmf = counts / N
cdf = np.cumsum(pmf)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.barplot(x=unique_values, y=pmf, palette="Blues", edgecolor="black")
plt.xlabel("Number of Customers")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF)")

plt.subplot(1, 2, 2)
plt.plot(unique_values, cdf, marker="o", linestyle="-", color="red")
plt.xlabel("Number of Customers")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

Q2-BINOMIAL PROBABILITY

```python
import scipy.stats as stats

N = 100
p = 0.7

expected_value = N * p
prob_60 = stats.binom.pmf(60, N, p)
prob_at_least_75 = 1 - stats.binom.cdf(74, N, p)

print(f"Expected number of students passing: {expected_value}")
print(f"Probability of exactly 60 students passing: {prob_60:.4f}")
print(f"Probability of at least 75 students passing: {prob_at_least_75:.4f}")
```
