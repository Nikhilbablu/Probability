```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
data = {'Advertising Budget': [1000, 2000, 3000, 4000, 1500, 2500, 3500, 1200, 4500, 5000],
        'Sales Revenue': [5000, 7000, 8000, 9500, 6000, 7500, 8700, 5500, 9900, 11000],
        'Social Media Engagement': [200, 250, 300, 400, 220, 280, 350, 210, 450, 500]}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(15, 5))
```

```python
plt.subplot(1, 3, 1)
sns.scatterplot(x='Advertising Budget', y='Sales Revenue', data=df)
plt.title('Advertising Budget vs. Sales Revenue')

plt.subplot(1, 3, 2)
sns.scatterplot(x='Social Media Engagement', y='Sales Revenue', data=df)
plt.title('Social Media Engagement vs. Sales Revenue')

plt.subplot(1, 3, 3)
sns.scatterplot(x='Advertising Budget', y='Social Media Engagement', data=df)
plt.title('Advertising Budget vs. Social Media Engagement')

plt.tight_layout()
plt.show()
```

```python
correlation_budget_sales, p_value_budget_sales = pearsonr(df['Advertising Budget'], df['Sales Revenue'])
correlation_engagement_sales, p_value_engagement_sales = pearsonr(df['Social Media Engagement'], df['Sales Revenue'])
correlation_budget_engagement, p_value_budget_engagement = pearsonr(df['Advertising Budget'], df['Social Media Engagement'])
```

```python
print(f"Pearson Correlation (Advertising Budget vs. Sales Revenue): {correlation_budget_sales:.2f}")
print(f"Pearson Correlation (Social Media Engagement vs. Sales Revenue): {correlation_engagement_sales:.2f}")
print(f"Pearson Correlation (Advertising Budget vs. Social Media Engagement): {correlation_budget_engagement:.2f}")
```

```python
print("\nDiscussion:")
print("Increasing the advertising budget shows a strong positive correlation with sales revenue. However, correlation doesn't imply causation.")
print("While the data suggests a positive relationship, other factors may influence sales.  It is not safe to conclude that increasing the budget *always* leads to higher revenue.")
print("Social Media Engagement also shows a strong positive correlation with Sales Revenue, suggesting that higher engagement might be related to higher sales.  Further analysis could investigate this relationship.")
```
